```python
import pandas as pd
```

## Prepare input data for classifiers

```python
pni_col_df = pd.read_csv('USE THIS_DESEQ2_PNI_0.001.csv')
pni_cols_list = list(pni_col_df.iloc[:,0])
df_pni = pd.read_csv('USE THIS_PNI_tpm_all.csv')
df_pni_found=df_pni[df_pni['X'].isin(pni_cols_list)]
df_pni_found.shape
```

```python
pni_found_list = list(df_pni_found['X'].unique())
```

```python
pni_not_found_list = [val for val in pni_cols_list if val not in pni_found_list]
```

```python
pni_not_found_split_list = [val.replace('.','-') for val in pni_not_found_list]
```

```python
df_pni_final = df_pni[df_pni['X'].isin(pni_cols_list + pni_not_found_split_list)]
```

```python
pni_found_list_final = list(df_pni_final['X'].unique())
```

```python
len(pni_found_list_final)
```

```python
df_pni_final.head()
```

```python
df_pni_final.drop(['Unnamed: 0'],axis =1, inplace=True)
```

```python
df_pni_final.head()
```

```python
df_pni_final.rename({'X':'gene'},axis = 1,inplace = True)
df_pni_final = df_pni_final.set_index(keys = ['gene'])
df_pni_final = df_pni_final.transpose()
df_pni_final= df_pni_final.reset_index()
df_pni_final.rename({'index':'ID'},axis = 1,inplace = True)
```

```python
df_pni_final.head()
```

```python
df_target = pd.read_csv('LVI_PNI_patient_information.csv')

df_target_pni = df_target[['ID','PNI']]

df_total_pni = pd.merge(left = df_pni_final, right = df_target_pni, on = 'ID', how = 'inner')

df_total_pni.to_csv('pni_001_cleaned.csv',index = False)
```
